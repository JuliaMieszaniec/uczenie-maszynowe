```python
x_train["income_cat"] = pd.cut(x_train["median_income"],
                                bins=[0., 1.5, 3.0, 4.5, 6., np.inf], 
                                labels=[1, 2, 3, 4, 5])
```

```python
x_train["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.xlabel("Kategoria dochodw")
plt.ylabel("Liczba dystryktw")
plt.show()
```

```python
df_train_rand_samp, df_test_rand_samp = train_test_split(
    x_train, test_size=0.2, random_state=42)
df_train, df_test = train_test_split(
    x_train, test_size=0.2, stratify=x_train["income_cat"], random_state=42)
```

```python
def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

compare_samp = pd.DataFrame({
    "cznie (%)": income_cat_proportions(x_train),
    "Warstwowe (%)": income_cat_proportions(df_test),
    "Losowe (%)": income_cat_proportions(df_test_rand_samp),
    }).sort_index()

compare_samp.index.name = "Kategoria dochodw"
compare_samp["Bd - warstwowe (%)"] = (compare_samp["Warstwowe (%)"] /
compare_samp["cznie (%)" ] - 1)
compare_samp["Bd - losowe (%)"] = (compare_samp["Losowe (%)"] /
compare_samp["cznie (%)"] - 1)
(compare_samp * 100).round(2)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 3), sharey=True)
x_train["population"].hist(ax=axs[0], bins=50)
x_train["population"].apply(np.log).hist(ax=axs[1], bins=50)
axs[0].set_xlabel("Population")
axs[1].set_xlabel("Log of population")
axs[0].set_ylabel("Number of districts")
plt.show()
```

```python
percentiles = [np.percentile(x_train["median_income"], p)
               for p in range(1, 100)]
x_train["median_income_percentiles"] = pd.cut(x_train["median_income"],
                                 bins=[-np.inf] + percentiles + [np.inf],
                                 labels=range(1, 100 + 1))
x_train["median_income_percentiles"] = x_train["median_income_percentiles"].astype(int)
x_train["median_income_percentiles"].hist(bins=50)
plt.xlabel("Median income percentile")
plt.ylabel("Number of districts")
plt.show()
```

```python
x_train["housing_median_age"].hist(bins=50)
```

```python
from sklearn.metrics.pairwise import rbf_kernel

x_train["age_simil_35_v1"] = rbf_kernel(
    x_train[["housing_median_age"]], [[35]], gamma=0.1)
x_train["age_simil_35_v2"] = rbf_kernel(
    x_train[["housing_median_age"]], [[35]], gamma=0.03)
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel("Mediana wieku domw")
ax1.set_ylabel("Liczba dystryktw")
ax1.hist(x_train["housing_median_age"], bins=50)
ax2 = ax1.twinx() # create a twin axis that shares the same x-axis
ax2.scatter(
x_train["housing_median_age"], x_train["age_simil_35_v1"],
color="green", label="gamma = 0,10")
ax2.scatter(
x_train["housing_median_age"], x_train["age_simil_35_v2"],
color="darkgreen", label="gamma = 0,03")
ax2.tick_params(axis='y', labelcolor="green")
ax2.set_ylabel("Podobiestwo wieku", color="green")
plt.legend(loc="upper left")
plt.show()
```
